## Step 8: Handle Missing Values

Sometimes datasets have missing values (NAN).
we will 
1. Count missing values in each column
2. Remove rows where a specific column is empty ("Cross Street")
3. View the updated dataset
4. These of columns has miss valuse (Mocodes, Vict Sex, Vict Descent, Premis Cd, Premsi Desc, Weapon Used Cd, Weapon Desc, Status, Crm cd 2, Cross Street)

```python
# count missing values in each column
print("Missing values per column:")
print(data_table.isnull().sum())
```

```python
# Remove rows where "Mocodes" is missing value.
data_table_cleaned  = data_table.dropna(subset = ["Cross Street"])

# view updated dataset 
data_table_cleaned.head()

# Remove rows where any of the listed columns are missing
data_table_cleaned = data_table.dropna(subset = ["Vict Sex","Vict Descent", "Premis Cd", "Mocodes"])
data_table_cleaned.isnull().sum()
```

```python
# Remove all rows that contain any NaN value
data_table_cleaned = data_table.dropna()
```

## Check missing values after cleaning

```python
print(data_table_cleaned.isnull().sum())

# Check number of rows before and after cleaning
print("Original dataset rows:", data_table.shape[0])
print("Cleaned dataset rows:", data_table_cleaned.shape[0])
```
